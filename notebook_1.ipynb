```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
trades = pd.read_csv(r'historical_data.csv')
sentiment = pd.read_csv(r'fear_greed_index.csv')
```

```python
# Step 2: Data Cleaning and Preparation

# Convert 'Timestamp IST' in trades DataFrame to datetime objects
trades['Timestamp IST'] = pd.to_datetime(trades['Timestamp IST'], errors='coerce')

# Convert 'date' in sentiment DataFrame to datetime objects
sentiment['date'] = pd.to_datetime(sentiment['date'], errors='coerce')

# Convert 'timestamp' in sentiment DataFrame to datetime objects
sentiment['timestamp_datetime'] = pd.to_datetime(sentiment['timestamp'], unit='s', errors='coerce')
```

```python
# after conversion
print("\nData types of trades after conversion:")
print(trades.dtypes)

print("\nData types of sentiment after conversion:")
print(sentiment.dtypes)
```

```python
# Check for missing values after conversion
print("Missing values in trades after date conversion:")
print(trades.isnull().sum())

print("\nMissing values in sentiment after date conversion:")
print(sentiment.isnull().sum())
```

```python
# Drop rows with missing 'Timestamp IST' values
trades_cleaned = trades.dropna(subset=['Timestamp IST']).copy()
```

```python
trades_cleaned.isnull().sum()
```

```python
print("\nInformation about the cleaned Trades DataFrame (trades_cleaned):")
trades_cleaned.info()

print("\nInformation about the cleaned Sentiment DataFrame:")
sentiment.info()
```

```python
#Feature Enginerring
#extract date and time from cleand trades data
trades_cleaned['Trade_Date'] = trades_cleaned['Timestamp IST'].dt.date
trades_cleaned['Trade_Time'] = trades_cleaned['Timestamp IST'].dt.time
trades_cleaned['Trade_Hour'] = trades_cleaned['Timestamp IST'].dt.hour
```

```python
sentiment = sentiment.sort_values(by='date')
```

```python
sentiment['sentiment_change'] = sentiment['value'].diff
```

```python

print("Sentiment DataFrame with daily sentiment change:")
display(sentiment.head())


print("\nTrades DataFrame with date and time features:")
display(trades_cleaned.head())
```

```python
# Data Merging

sentiment['Sentiment_Date'] = sentiment['date'].dt.date
```

```python
#we merge the trades_cleaned and sentiment dataframes based on the date
merged_df = pd.merge(trades_cleaned, sentiment[['Sentiment_Date', 'value', 'classification', 'sentiment_change']],
                     left_on='Trade_Date', right_on='Sentiment_Date', how='left')
```

```python
# we Drop the extra date column from the sentiment dataframe after merging
merged_df = merged_df.drop(columns=['Sentiment_Date'])
```

```python
# Display the first 5 rows of merged data frame
print("Merged DataFrame:")
display(merged_df.head())
```

```python
print("Information about Merged DataFrame:")
merged_df.info()
```

```python
#EDA
print("First 5 rows of the merged DataFrame:")
display(merged_df.head())
```

```python
print("Information about the merged DataFrame:")
merged_df.info()
```

```python
print("Statistics of the merged DataFrame:")
display(merged_df.describe())
```

```python
# relationship between sentiment value and Closed PnL
plt.figure(figsize=(10, 4))
sns.scatterplot(x='value', y='Closed PnL', data=merged_df, alpha=0.5)
plt.title('Closed PnL vs. Sentiment Value')
plt.xlabel('Sentiment Value')
plt.ylabel('Closed PnL')
plt.show()
```

```python
# relationship between average Closed PnL by sentiment classification
avg_pnl_by_sentiment = merged_df.groupby('classification')['Closed PnL'].mean().reset_index()
plt.figure(figsize=(10, 4))
sns.barplot(x='classification', y='Closed PnL', data=avg_pnl_by_sentiment)
plt.title('Average Closed PnL by Sentiment Classification')
plt.xlabel('Sentiment Classification')
plt.ylabel('Average Closed PnL')
plt.show()
```

```python
#relationship between Execution Price and Closed PnL
plt.figure(figsize=(10,4))
sns.scatterplot(x='Execution Price', y='Closed PnL', data=merged_df, alpha=0.5)
plt.title('Closed PnL vs. Execution Price')
plt.xlabel('Execution Price')
plt.ylabel('Closed PnL')
plt.show()
```

```python
#relationship between Size USD and Closed PnL
plt.figure(figsize=(10,5))
sns.scatterplot(x='Size USD', y='Closed PnL', data=merged_df, alpha=0.5)
plt.title('Closed PnL vs. Size USD')
plt.xlabel('Size USD')
plt.ylabel('Closed PnL')
plt.show()
```

```python
#Analyze the average Closed PnL by Side
avg_pnl_by_side = merged_df.groupby('Side')['Closed PnL'].mean().reset_index()

plt.figure(figsize=(8, 4))
sns.barplot(x='Side', y='Closed PnL', data=avg_pnl_by_side)
plt.title('Average Closed PnL by Side')
plt.xlabel('Side')
plt.ylabel('Average Closed PnL')
plt.show()
```

```python
# Analyze the average Closed PnL by Direction
avg_pnl_by_direction = merged_df.groupby('Direction')['Closed PnL'].mean().reset_index()

plt.figure(figsize=(10, 4))
sns.barplot(x='Direction', y='Closed PnL', data=avg_pnl_by_direction)
plt.title('Average Closed PnL by Direction')
plt.xlabel('Direction')
plt.ylabel('Average Closed PnL')
plt.xticks(rotation=90)
plt.show()
```

```python
# Aaverage Closed PnL by Coin
avg_pnl_by_coin = merged_df.groupby('Coin')['Closed PnL'].mean().reset_index().sort_values(by='Closed PnL', ascending=False)

plt.figure(figsize=(11, 5))
sns.barplot(x='Coin', y='Closed PnL', data=avg_pnl_by_coin.head(15)) # Displaying top 15 for clarity
plt.title('Average Closed PnL by Coin (Top 15)')
plt.xlabel('Coin')
plt.ylabel('Average Closed PnL')
plt.xticks(rotation=90)
plt.show()
```

```python
#Top 5 traded coins
top_coins = merged_df['Coin'].value_counts().head(5).index.tolist()
print(f"Top 5 most traded coins: {top_coins}")
```

```python
#relationship between sentiment and Closed PnL for top coins
for coin in top_coins:
    coin_df = merged_df[merged_df['Coin'] == coin]
    if not coin_df.empty:
        avg_pnl_by_sentiment_coin = coin_df.groupby('classification')['Closed PnL'].mean().reset_index()

        print(f"\nAverage Closed PnL by Sentiment Classification for {coin}:")
        display(avg_pnl_by_sentiment_coin)
```

```python
# relationship between sentiment and Closed PnL for top coins
for coin in top_coins:
    coin_df = merged_df[merged_df['Coin'] == coin]
    if not coin_df.empty:
        avg_pnl_by_sentiment_coin = coin_df.groupby('classification')['Closed PnL'].mean().reset_index()

        plt.figure(figsize=(10,4))
        sns.barplot(x='classification', y='Closed PnL', data=avg_pnl_by_sentiment_coin)
        plt.title(f'Average Closed PnL by Sentiment Classification for {coin}')
        plt.xlabel('Sentiment Classification')
        plt.ylabel('Average Closed PnL')
        plt.show()
```

## Summary of EDA Findings

Based on the exploratory data analysis performed on the merged historical trader and market sentiment data, here are the key observations:

**General Characteristics of the Merged Dataset:**
- The merged dataset contains trading data enriched with daily sentiment scores and classifications.
- There were a significant number of trades where the timestamp could not be parsed, resulting in missing sentiment data for those specific trades after merging.
- The dataset covers a range of execution prices, trade sizes (in tokens and USD), and profit/loss outcomes.

**Relationships between Closed PnL and Sentiment:**
- The scatter plot of `Closed PnL` vs. `Sentiment Value` did not show a clear linear correlation, suggesting that a high or low sentiment value alone does not directly guarantee a proportional profit or loss.
- However, when analyzing the average `Closed PnL` by `Sentiment Classification`, we observed variations. For instance, 'Extreme Greed' and 'Fear' sentiment classifications appear to be associated with higher average Closed PnL compared to 'Neutral' or 'Extreme Fear' based on the bar plot. It's important to note that 'Extreme Greed' also shows a high standard deviation, indicating potential for both high profits and losses.

**Relationships between Closed PnL and other Trading Attributes:**
- The scatter plots of `Closed PnL` vs. `Execution Price` and `Size USD` also did not show strong linear relationships. This suggests that simply trading at a high execution price or with a large size in USD does not directly correlate with higher PnL across all trades.
- Analyzing average `Closed PnL` by `Side` showed slightly higher average PnL for 'SELL' trades compared to 'BUY' trades.
- Analyzing average `Closed PnL` by `Direction` provided more granular insights. 'Close Short' and 'Sell' directions had the highest average Closed PnL, while 'Short > Long' had a negative average PnL. 'Buy', 'Open Long', 'Open Short', and 'Spot Dust Conversion' had average PnL close to zero.

**Insights from Analyzing Trader Performance by Coin:**
- The average `Closed PnL` varies significantly across different `Coin` types. Some coins (e.g., SEI, USUAL, ENA) show considerably higher average PnL than others, including major coins like BTC and ETH, and some coins show negative average PnL (e.g., FARTCOIN). This highlights the importance of coin selection in trading outcomes.
- The relationship between sentiment and `Closed PnL` for top traded coins (`HYPE`, `@107`, `BTC`, `ETH`, `SOL`) appears to be coin-specific. For example:
    - For `HYPE` and `@107`, 'Greed' and 'Extreme Greed' sentiments were associated with higher average PnL.
    - For `BTC`, 'Fear' sentiment had the highest average PnL.
    - For `ETH`, 'Fear' sentiment was associated with a positive average PnL, while 'Extreme Greed' and 'Greed' were associated with negative average PnL.
    - For `SOL`, 'Fear' and 'Greed' sentiments were associated with higher average PnL.

These findings suggest that while overall market sentiment might play a role, its impact on trading performance is not uniform and can be influenced by the specific cryptocurrency being traded and the trading direction. Further analysis could delve deeper into these coin-specific and direction-specific patterns.
